```python
import matplotlib.pyplot as plt
import pandas as od
import numpy as np
```

```python
dev_age=[20,22,21,23,47,36,46,55]
dev_main =[10500,15000,12500,19000,48000,36000,45000,52000]
```

```python
plt.plot(dev_age,dev_main)
plt.title('salary of dev')
plt.xlabel("age")
plt.ylabel("salary")
```

```python
dev_age=[20,22,21,23,47,36,46,55]
dev_main =[10500,15000,12500,19000,48000,36000,45000,52000]
plt.plot(dev_age,dev_main,label='main_dev')

py_dev=[10000,13000,12000,14500,45000,38500,40000,50000]
plt.plot(dev_age,py_dev,label='py_dev')

plt.title('salary of dev')
plt.xlabel("age")
plt.ylabel("salary")
plt.legend()
plt.show()
```

```python
dev_age=[20,22,21,23,47,36,46,55]
dev_main =[10500,15000,12500,19000,48000,36000,45000,52000]
plt.plot(dev_age,dev_main,label='main_dev',color='red')

py_dev=[10000,13000,12000,14500,45000,38500,40000,50000]
plt.plot(dev_age,py_dev,label='py_dev',color='green',alpha=0.6)

plt.title('salary of dev')
plt.xlabel("age")
plt.ylabel("salary")
plt.legend()
plt.grid()
plt.show()
```
